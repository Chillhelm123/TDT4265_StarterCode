### Task 3c)
The use of different aspect ratios gives the classification step a starting point (anchor). The starting point uses prior knowledge about the aspect ratio the object is most likely to occur as, which could make the object detection/classification faster. For example a bounding box for a human is more often taller than it is wide.

### Task 3d)
The main difference between YOLO v1/v2 and SSD is that YOLO divides the input image into a grid and uses a fixed set of bounding boxes per grid cell determined by k-means clustering on the training dataset, while SSD uses predefined anchors at every spatial location in the feature map using different scales and aspect ratios. YOLO is also faster than SSD, but less likely to detect smaller objects, due to the fixed bounding boxes.
